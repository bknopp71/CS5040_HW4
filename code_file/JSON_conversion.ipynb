```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import os

# 🔧 Set your folder path
base_path = "/content/drive/MyDrive/silver_detection/cleaned_dataset/"

# CSV files to convert
csv_files = [
    "cleaned_non_silver_metadata.csv",
    "cleaned_silver_metadata.csv"
]

# Convert each CSV to JSON
for file in csv_files:
    csv_path = os.path.join(base_path, file)
    json_path = os.path.join(base_path, file.replace(".csv", ".json"))

    # Read CSV
    df = pd.read_csv(csv_path)

    # Optional: clean missing values
    df = df.fillna("null")

    # Save as JSON (record format = best for ML)
    df.to_json(json_path, orient="records", indent=4)

    print(f"✅ Converted: {file} → {json_path}")

print("🚀 All files converted successfully!")
```
